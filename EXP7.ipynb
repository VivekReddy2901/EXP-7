```python
# Factorial of a number

n = int(input("Enter N : "))
i = 1
while (n > 0) :
  i = i * n
  n = n - 1
print("Factorial :",i)
```

```python
# Fibonacci series using while loop

n = int(input("Enter Number of terms :"))
i = 1
a = 0
b = 1
while (i <= n):
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i = i + 1
```

```python
# Reverse a number
num = int(input("Enter a number: "))
rev = 0
while num > 0 :
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print("reversed number :",rev)
```

```python
# check palindrone number

num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0 :
  rev = rev * 10 + num % 10
  num //= 10
if temp == rev :
  print("Palindrone")
else:
  print("Not Palindrone")
```

```python
# check palindrone number

st = input("Enter the string: ")
rev = st[::-1]
if st == rev :
  print("Palindrone")
else :
  print("Not palindrone")
```

```python
# check palindrome string

s  = "vivek"
i,j = 0 , len(s) - 1
is_palindrome = True
while i < j :
  if s[i] != s[j]:
    is_palindrome = False
    break
  i = i + 1
  j = j - 1
if is_palindrome :
  print("yes")
else :
  print("no")
```

```python
# count digits in number

num = int(input("Enter a number: "))
count = 0

while num > 0:
  count = count + 1
  num = num// 10

print("Number of digits: ",count)
```

```python
# search element in a list

nums = [10, 20, 30, 40, 50]
key = int(input("Enter the element to search: "))
i = 0

while i < len(nums) :
  if nums[i] == key :
    print("Element found at index", i)
    break
  i = i + 1
else:
  print("Element not found")
```
